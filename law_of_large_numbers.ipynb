#큰수의 법칙(Law of Large Numbers)

### 표본의 수가 커질 수록 표본의 통계랑은 점점 모집단의 모수와 같아진다(가까워진다.)

```python
#라이브러리 로드
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#모집단 생성

#평균이 100, 표준편차 20, 10000개의 모집단
population = np.random.normal(100, 20, 10000)
print("모집단의 크기 : ", population)
print("모평균 : ", np.mean(population))
print("모분산 : ", np.var(population))
```

## 1. 모평균 확인

표본의 크기(n)를 증가시키고 그래프를 그려 확인  
표본은 1개에서 3000개로 증가시킴

```python


sample_means = []
sample_vars = []
for n in range(1, 3001):
  sample = np.random.choice(population, n)
  sample_means.append(np.mean(sample))
  sample_vars.append(np.var(sample))

print("sample_means : ",len(sample_means))
print("sample_vars : ", len(sample_vars))
```

그래프를 그려 표본의 크기(n)가 커질 수록  
모평균과 가까워 지는지 확인

```python
plt.figure(figsize = (8,8))
ax = sns.lineplot(data=pd.DataFrame(sample_means), palette='Blues', legend=False)
#모평균
ax.axhline(y=np.mean(population), color='red')
plt.show()

print("표본의 크기가 10개일때 평균 : ", sample_means[9])
print("표본의 크기가 3000개일때 평균 : ", sample_means[2999])
print("모평균 : ", np.mean(population))
```

표본의 크기(n)가 증가할수록 점점 모평균에 가까워지는 것을 볼 수 있다.

##2. 모분산 확인

방식은 위와 같음

```python
plt.figure(figsize = (8,8))
ax = sns.lineplot(data=pd.DataFrame(sample_vars), palette='Blues', legend=False)
#모평균
ax.axhline(y=np.var(population), color='red')
plt.show()

print("표본의 크기가 10개일때 분산 : ", sample_vars[9])
print("표본의 크기가 3000개일때 분산 : ", sample_vars[2999])
print("모분산 : ", np.var(population))
```

마찬가지로 표본의 크기(n)가 증가할수록 점점 모분산에 가까워지는 것을 확인할 수 있음
